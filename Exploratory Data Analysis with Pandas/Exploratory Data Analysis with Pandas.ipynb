# Exploratory Data Analysis with Pandas

#### Memanggil library di Python

```python
import pandas as pd
```

#### Membaca file CSV sebagai data frame

```python
order = pd.read_csv('order.csv')
order
```

#### Melihat struktur baris dan kolom dari data frame

```python
order.shape
```

```python
order.tail(6)
```

Menampilkan sebagian kolom (kolom 'price' dan 'product_category_name') dari konten teratas pada data frame.

#### Statistik deskriptif

Secara umum function describe() secara otomatis memberikan summary statistik untuk kolom numerik dan mengabaikan kolom category.

```python
order.describe()
```

Menampilkan summary statistik untuk kolom category.

```python
order.describe(include = 'object')
```

Menampilkan summary statistik dari semua kolom (kolom numerik dan category).

```python
order.describe(include = 'all')
```

```python
print(f"mean price = {order.loc[:, 'price'].mean()}")
```

```python
print(f"median price = {order.loc[:, 'price'].median()}")
```

```python
print(f"mode price =\n{(order.loc[:, 'price'].mode())}")
```

#### Membuat histogram

```python
order['price'].plot(figsize = (4, 5), kind = 'hist', bins = 15)
```

#### Standar deviasi

```python
print(f"standar deviasi product weight gram = {order.loc[:, 'product_weight_gram'].std()}")
```

#### Varians

```python
print(f"varians product weight gram = {order.loc[:, 'product_weight_gram'].var()}")
```

#### Interquartile range (IQR)

```python
q1 = order['product_weight_gram'].quantile(0.25)
q3 = order['product_weight_gram'].quantile(0.75)
iqr = q3 - q1
print(f'product weight gram:\nQ1 = {q1}, Q3 ={q3}, dan IQR ={iqr}')
```

#### Rename kolom data frame

Melakukan rename kolom dari 'freight_value' menjadi 'shipping_cost'.

```python
order.rename(columns = {'freight_value': 'shipping_cost'}, inplace = True)
order
```

#### Groupby

```python
order['price'].groupby(order['payment_type']).mean()
```

Menggunakan kolom 'payment_type' sebagai aggregate dan kemudian menghitung mean dari kolom 'price' pada setiap aggregate tersebut.

#### Sorting

Secara ascending.

```python
order.sort_values(by = 'price',)
```

Secara descending.

```python
order.sort_values(by = 'price', ascending = False)
```
